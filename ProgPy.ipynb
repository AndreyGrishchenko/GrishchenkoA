```python
from numpy import random
from scipy import linalg
import numpy as np
import matplotlib.pyplot as plt

CondArr = []
V = np.arange(1, 300)

def cond(X):
    
    inv_matrix=linalg.inv(X)
    norm1=linalg.norm(inv_matrix,1)
    norm2=linalg.norm(X,1)
    
    return(norm1*norm2)

for n in V:
  Val=random.randn(n,n)
  CondArr.append(cond(Val))
plt.loglog(V,CondArr,'.')
```

На данный момент смог построить график, который отображает зависисмость размера матрицы от числа обусловленности.
