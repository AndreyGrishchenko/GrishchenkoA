```python
import numpy as np
from numpy import random
from scipy import linalg
import matplotlib.pyplot as plt

CondArr = []
V = np.arange(1, 300)
c1=4
c2=0.25
m=1.8
s1=c1*(V**m)
s2=c2*(V**m)

def cond(X):
    
    inv_matrix=linalg.inv(X)
    norm1=linalg.norm(inv_matrix,1)
    norm2=linalg.norm(X,1)
    
    return(norm1*norm2)

for n in V:
  Val=random.randn(n,n)
  CondArr.append(cond(Val))
plt.loglog(V,CondArr,'.')
plt.loglog(V,s1)
plt.loglog(V,s2)
```

Изучение зависимости роста числа обусловленности случайной матрицы от её порядка.

Для этого построим график, который отображает зависимость числа обусловленности матрицы от её порядка. Элементы матрицы распределены по стандартному распределению Гаусса. Также построим 2 прямые, которые являются графиками среднего роста cond1A(n). Подберём такие коэффициенты c1,c2 и m, чтобы выполнялось уравнение c1*n^m<=cond1A(n)<=c2*n^m. Из графика видно, чтобольшинство точек находятся между прямыми s1 и s2.
Таким образом, мы выяснили, что в среднем cond1A(n) прямо пропорциональна показательной функции от n.
